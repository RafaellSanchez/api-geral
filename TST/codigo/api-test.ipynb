```python
import requests

url = 'https://opendata.api.bb.com.br/open-banking/channels/v1/banking-agents?page=1&page-size=2'

response = requests.get(url)
dados = response.json()
print(dados)

if 'data' in dados:
    data = dados['data']
    if 'brand' in data:
        brand_data = data['brand']
        if 'companies' in brand_data:
            companies = brand_data['companies']
            for company_info in companies:
                print("Company Name:", company_info.get('name'))
                banking_agents = company_info.get('bankingAgents', [])
                for agent_info in banking_agents:
                    print("Agent Name:", agent_info.get('identification', {}).get('corporationName'))
                    locations = agent_info.get('locations', [])
                    if locations:
                        print("Address:", locations[0].get('postalAddress', {}).get('address'))
                    else:
                        print("Address: Not available")
                    print("Availability:", agent_info.get('availability', {}).get('exception', 'Unknown'))
                    print()
        else:
            print("No 'companies' data available")
    else:
        print("No 'brand' data available")
else:
    print("No 'data' key found in response")
```

```python
import requests
import pandas as pd

url = 'https://opendata.api.bb.com.br/open-banking/channels/v1/banking-agents?page=1&page-size=2'

response = requests.get(url)
dados = response.json()

print(dados)  # Imprime os dados retornados pela API

agent_list = []

if 'brand' in dados:
    brand_data = dados['brand']
    if 'companies' in brand_data:
        companies = brand_data['companies']
        for company_info in companies:
            company_name = company_info.get('corporateName')
            banking_agents = company_info.get('agents', [])
            for agent_info in banking_agents:
                agent_name = agent_info.get('name')
                address = agent_info.get('address')
                availability = agent_info.get('availability', 'Unknown')
                agent_list.append({'Company Name': company_name, 'Agent Name': agent_name, 'Address': address, 'Availability': availability})
    else:
        print("No 'companies' data available")
else:
    print("No 'brand' data available")

# Convertendo para DataFrame
df = pd.DataFrame(agent_list)
print(df)
```
