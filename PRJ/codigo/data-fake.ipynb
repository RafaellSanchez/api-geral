```python
from faker import Faker
import pandas as pd

fake = Faker('pt_BR')

dados_list = []

while len(dados_list) < 10:  # Garante que serão gerados 10 dados brasileiros
    nome = fake.name()
    pais = fake.country()
    
    # Verifica se o país gerado é o Brasil
    if pais == 'Brasil':
        endereco = fake.address()
        email = fake.email()
        numero_telefone = fake.phone_number()
        
        dados_list.append({
            'nome': nome,
            'email': email,
            'telefone': numero_telefone,
            'endereco': endereco,
            'pais': pais
        })

df = pd.DataFrame(dados_list)
display(df)
```
