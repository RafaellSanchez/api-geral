```python
import pandas as pd
import requests
```

```python
url = 'https://parallelum.com.br/fipe/api/v1/carros/marcas'
response = requests.get(url)

dados = response.json()

data_list = []

if dados:
    if 'marcas' in dados:
        modelos = dados['marcas']
        for modelo in modelos:
            codigo = modelo['codigo']
            nome = modelo['nome']
            # print(f"codigo: {codigo}")
            # print(f"nome: {nome}")
            
            data_list.append({
                'codigo': codigo,
                'modelo': nome
            })
    else:  
        for modelo in dados:
            codigo = modelo['codigo']
            nome = modelo['nome']
            # print(f"codigo: {codigo}")
            # print(f"nome: {nome}")
            
            data_list.append({
                'codigo': codigo,
                'modelo': nome 
            })

df = pd.DataFrame(data_list)
display(df)

# file = 'marcas.txt'
# path = '/workspaces/api-geral/CAR/dados/'

# df.to_csv(f'{path}{file}', sep=';', index=False)
# print(f'Arquivo salvo: {file}')
# print(f'Caminho: {path}')
```

```python
import sqlite3
```

```python
file = '/workspaces/api-geral/CAR/sql/fipe.db'

conn = sqlite3.connect(file)
cursor = conn.cursor()
```

```python
show_tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
cursor.execute(show_tables_query)

# Recuperar o resultado da consulta e imprimir as tabelas
tables = cursor.fetchall()
print("Tabelas no banco de dados:")
for table in tables:
    print(table[0])
```

```python
tabela = 'tb_ford'

# Execute uma consulta para recuperar os nomes das colunas da tabela
cursor.execute("PRAGMA table_info({})".format(tabela))
colunas = cursor.fetchall()

for coluna in colunas:
    print(coluna[1])

# cursor.close()
# conn.close()
```

```python
query = 'select * from tb_ford'
cursor.execute(query)

resultados = cursor.fetchall()

# Exibir os resultados
for linha in resultados:
    print(linha)
```

```python
from pyspark.sql import SparkSession
```

```python
# Inicializa uma sessão do Spark
spark = SparkSession.builder \
    .appName("Leitura") \
    .getOrCreate()
```

```python
url_db = '/workspaces/api-geral/CAR/sql/fipe.db'
query = 'select * from tb_ford'

df = spark.read \
    .format("jdbc") \
    .option("url", url_db) \
    .option("dbtable", query) \
    .load()

# Mostre os dados do DataFrame
df.show()
```
