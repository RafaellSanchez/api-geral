```python
import pandas as pd
import json
```

```python
file = '/workspaces/api-geral/CAR/repo_ford/ford/modelo_ford.txt'
```

```python
df = pd.read_csv(file, sep=';')
display(df)
```

```python
df = pd.DataFrame(df)
```

```python
# Corrigir o formato do JSON para garantir que as chaves estejam entre aspas duplas
df['modelo'] = df['modelo'].str.replace("'", '"')

# Converter a coluna 'modelo' de strings JSON para dicionários
df['modelo'] = df['modelo'].apply(json.loads)

# Extrair as chaves do dicionário na coluna 'modelo'
df['codigo'] = df['modelo'].apply(lambda x: x['codigo'])
df['nome'] = df['modelo'].apply(lambda x: x['nome'])
# Remover a coluna 'modelo' se não for mais necessária
df.drop(columns=['modelo'], inplace=True)
display(df)
```

```python
df['id'] = 22
display(df)
```

```python
caminho = '/workspaces/api-geral/CAR/repo_ford/ford/'
arquivo = 'modelo_ford_atlz.txt'
df = df.to_csv(f'{caminho}{arquivo}', sep= ';', index=False)
print(f'Arquivo salvo: {arquivo}')
print(f'Diretório: {caminho}')
```
