```python
create_gold_sp = '''
CREATE TABLE IF NOT EXISTS sp_dados_geo (
    uuid TEXT,
    id INTEGER,
    nome TEXT,
    idMunicipio INTEGER,
    nomeMunicipio TEXT,
    idMacro INTEGER,
    nomeMacroMunicipio TEXT,
    idUf INTEGER,
    estado TEXT,
    siglaUf TEXT,
    idRegiao INTEGER,
    regiao TEXT,
    siglaRegiao TEXT,
    dtIgtao DATE
);
'''

# Executar o script para criar a nova tabela
cursor.execute(create_gold_sp)

# Consulta para selecionar os dados da tabela bronze_dados_geo e inseri-los na tabela silver_dados_geo
consulta_insert = '''
INSERT INTO sp_dados_geo (
    uuid,
    id,
    nome,
    idMunicipio,
    nomeMunicipio,
    idMacro,
    nomeMacroMunicipio,
    idUf,
    estado,
    siglaUf,
    idRegiao,
    regiao,
    siglaRegiao,
    dtIgtao
)
SELECT
    uuid,
    id,
    nome,
    idMunicipio,
    nomeMunicipio,
    idMacro,
    nomeMacroMunicipio,
    idUf,
    estado,
    siglaUf,
    idRegiao,
    regiao,
    siglaRegiao,
    strftime('%Y-%m-%d', 'now') AS dtIgtao
FROM
    silver_dados_geo
WHERE estado = 'São Paulo'    
;
'''

# Executar a consulta para selecionar e inserir os dados
cursor.execute(consulta_insert)

# Commit para salvar as alterações
conn.commit()

# Fechar a conexão
conn.close()
```
