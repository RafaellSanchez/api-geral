```python
import pandas 
import os
import re
import shutil
```

```python
# Definir o caminho e a palavra-chave
caminho = '/workspaces/api-geral/GEO/CEP/dados/'
palavra_chave = 'mairipora'

# Listar todos os arquivos no diretório
lista_caminho = os.listdir(caminho)

# Lista para armazenar os DataFrames de cada arquivo
dfs = []

# Iterar sobre os arquivos
for nome_arquivo in lista_caminho:
    # Verificar se a palavra-chave está presente no nome do arquivo
    if palavra_chave in nome_arquivo:
        src_arquivo = os.path.join(caminho, nome_arquivo)
        print("Lendo arquivo:", src_arquivo)
        
        # Ler o arquivo CSV como DataFrame
        df = pd.read_csv(src_arquivo, delimiter=';')
        # Adicionar o DataFrame à lista
        dfs.append(df)

# Se houver algum DataFrame na lista
if dfs:
    # Concatenar todos os DataFrames da lista em um único DataFrame
    df_final = pd.concat(dfs)
    
    # Salvar o DataFrame final em um arquivo CSV
    caminho_saida = '/workspaces/api-geral/GEO/CEP/dados/resultado.txt'
    df_final.to_csv(caminho_saida, sep=';', index=False)
    print("Arquivo final salvo em:", caminho_saida)
else:
    print("Nenhum arquivo encontrado com a palavra-chave especificada.")
```
