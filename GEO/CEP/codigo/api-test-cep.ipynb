```python
url = 'https://viacep.com.br/ws/SP/Mairipora/Mairiporã/json/'

response = requests.get(url)
print('realizando response!')
if response.status_code == 200:
    print ('OK!')
else:
    print ('Erro!')

dados = response.json()
print('transformando no formato json')

lista = []

for resultado in dados:
    cep = resultado['cep']
    logradouro = resultado['logradouro']
    bairro = resultado['bairro']
    localidade = resultado['localidade']
    uf = resultado['uf']
    ibge = resultado['ibge']
    gia = resultado['gia']
    ddd = resultado['ddd']
    siafi = resultado['siafi']
    
    lista.append({
        'cep': cep,
        'logradouro': logradouro,
        'bairro': bairro,
        'localidade': localidade,
        'uf': uf,
        'ibge': ibge,
        'gia': gia,
        'ddd': ddd,
        'siafi': siafi
        
        })
    
df = pd.DataFrame(lista)
display(df)

caminho = '/workspaces/api-geral/GEO/CEP/dados/'
arquivo = 'dados_sp_mairipora.txt'

df.to_csv(f'{caminho}{arquivo}', sep=';', index=False)
print('arquivo salvo')

    
```

```python
import requests
from datetime import datetime


timestamp = datetime.now().strftime('%Y-%m-%d')

url = 'https://viacep.com.br/ws/07600740/json/'

response = requests.get(url)
print('realizando response!')
if response.status_code == 200:
    print('OK!')
    dados = response.json()
    print('transformando no formato json')
    
    caminho = '/workspaces/api-geral/GEO/CEP/dados/'
    arquivo = f'dados_cep_{timestamp}.txt'

    num_informacoes = 0

    with open(f'{caminho}{arquivo}', 'w') as f:
        for chave, valor in dados.items():
            if isinstance(valor, dict):  # Verifica se o valor é um dicionário
                for sub_chave, sub_valor in valor.items():
                    print(f'{sub_chave}: {sub_valor}', file=f)
                    num_informacoes += 1
            else:
                print(f'{chave}: {valor}', file=f)
                num_informacoes += 1
    
    print(f'Foram encontradas {num_informacoes} informações e salvas no arquivo {arquivo}.')
else:
    print('Erro!')
```
